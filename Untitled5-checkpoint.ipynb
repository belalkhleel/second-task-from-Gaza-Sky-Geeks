```python
df.drop(columns=['new_deaths_smoothed','new_cases_per_million','total_cases_per_million'],axis=0)
```

```python
df.rename(columns = {'date':'Date',"location":"Country","continent":"Continent","iso_code":"ISO_code"}, inplace = True)
df.head(n=5)
```

```python
lst=[] 
lst=(df['Continent'].unique())
print(lst)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
print(df2)
```

```python
df3 = df2.groupby(['Date', 'Country'])[['total_cases', 'total_deaths', 'total_vaccinations']].sum().reset_index()
print(df3)
```

```python
df["total_cases"] = df["total_cases"].fillna(0)
df["total_deaths"] = df["total_deaths"].fillna(0)
df["total_vaccinations"] = df["total_vaccinations"].fillna(0)
df.head(10)
```

```python
lst2=(df3.loc[df["total_deaths"]>=1000000,["Country"]])
print(lst2)
```
